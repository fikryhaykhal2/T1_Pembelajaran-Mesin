Membaca file CSV yang berisi tentang data kanker paru-paru dan disiman pada variable cancer


```python
cancer = pd.read_csv('/content/survey lung cancer.csv')
```

Baris ini mengganti nilai yang awalnya kategori menjadi numerik agar lebih mudah di proses

```python
cancer['GENDER'] = cancer['GENDER'].map({'F': 0, 'M': 1})
cancer['LUNG_CANCER'] = cancer['LUNG_CANCER'].map({'NO': 0, 'YES': 1})

cancer.head()
```

Baris ini bertujuan untuk membagi data menjadi atribut (X) dan label (y) untuk tujuan pemodelan

```python
X = cancer.drop('LUNG_CANCER', axis=1)
y = cancer['LUNG_CANCER']
```

Baris ini ini memungkinkan kita untuk membagi dataset menjadi set pelatihan dan pengujian dengan proporsi tertentu, yaitu test size/data uji sebesar 0.33 atau 33% dan 67% akan digunakan untuk pelatihan. Serta nilai random state 42 ntuk mengontrol randomness saat pembagian dataset.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
